```python
import torchvision.transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torch, os, re
import pandas as pd
from torchvision.io import read_image
from datetime import datetime
import torch.nn as nn
import numpy as np
import torch, json
import torchvision
import seaborn as sns
ROOM = 'living_room'
gradcam_table = pd.read_csv('/home/klimenko/nwl/new_living_visual/Methodology/CLASSIFICATION/results/cities_small3_kitchen/gradcam_results.csv')
gradcam_table_2 = gradcam_table.iloc[:, 2:156]
gradcam_table_3 = gradcam_table_2.drop(gradcam_table_2.columns[1:5], axis=1)
sum_df = gradcam_table_3.groupby('label').sum()
sum_df2 = sum_df.apply(lambda row: row[0:] / np.linalg.norm(row[0:]), axis=1) # normalize each row
sum_df2 = sum_df2.apply(lambda x: x ** 1)
sum_df2.columns


column_sums = sum_df2.sum()
sorted_columns = column_sums.sort_values(ascending=False)
ordered_columns = sorted_columns.index.tolist()
columns_to_preserve = ordered_columns[0:40]
sum_df2 = sum_df2.loc[:, columns_to_preserve]


print(ordered_columns[0:10])
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(sum_df2, cmap='YlOrBr')
plt.title('GradCAM attention to object classes, '+ROOM)
# Show the heatmap
plt.show()
sum_df44 =sum_df2
```

```python
from scipy.stats import levene

variances_A = sum_df44.var(axis=1)
variances_B = sum_df22.var(axis=1)

# Perform Levene's test
statistic, p_value = levene(variances_A, variances_B)

# Set the significance level (e.g., 0.05)
alpha = 0.05

# Compare the p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis. Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. Variances are not significantly different.")
```

```python
vectorlist=[]
labellist=[]
for row in sum_df22.iterrows():
    vectorlist.append(list(row[1]))
    labellist.append('least characteristic')
for row in sum_df44.iterrows():
    vectorlist.append(list(row[1]))
    labellist.append('most characteristic')
    
import sklearn.decomposition # PCA
import sklearn.manifold # MDS, t-SNE

cmap = plt.get_cmap('Reds')
colors = {'most characteristic':'blue',
          'least characteristic':'hotpink',
         }

tree = np.float64(np.asarray(vectorlist))


tsne_operator = sklearn.manifold.TSNE(n_components=2)
tree_tsne = tsne_operator.fit_transform(tree)
c = [colors[k] for k in labellist]

import matplotlib.pyplot as plt
plt.figure(figsize=(3, 3))
plt.scatter(list(tree_tsne[:,0]), list(tree_tsne[:,1]), c=c,cmap=cmap, s=145)
unique_colors = list(colors.values())
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=8) for color in unique_colors]
legend_labels = [city for city, color in colors.items() if color in unique_colors]  # City names corresponding to the unique colors
plt.subplots_adjust(right=0.75)
legend = plt.legend(legend_elements, legend_labels, title='Cities', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title('TSNE embeddings of cities '+ROOM+', RGB from ResNet-50')
#plt.show()
```
